```python
# Install & Import
!pip install imbalanced-learn openpyxl

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from collections import Counter
from google.colab import files

# Upload file
uploaded = files.upload()
nama_file = list(uploaded.keys())[0]

# Load data
df = pd.read_excel(nama_file)
df = df[pd.to_numeric(df['No'], errors='coerce').notna()]
df['No'] = df['No'].astype(int)

# Encode fitur
fitur_kategorik = [
    'Frekuensi Membaca', 'Jumlah Bahan Bacaan',
    'Durasi Membaca (Jam/Hari)', 'Frekuensi Akses Internet',
    'Durasi Akses Internet (Jam/Hari)'
]
fitur_numerik = ['Latitude', 'Longtitude', 'Tingkat Kegemaran Membaca']

le = LabelEncoder()
df_enc = df.copy()
for col in fitur_kategorik:
    df_enc[col] = le.fit_transform(df_enc[col].astype(str))

X = df_enc[fitur_numerik + fitur_kategorik]
y = le.fit_transform(df_enc['Kategori Penilaian'])

# SMOTE
smote = SMOTE(random_state=42, k_neighbors=3)
X_res, y_res = smote.fit_resample(X, y)
print("Setelah SMOTE:", Counter(y_res))

# Split training & testing
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42
)

print("Jumlah data training:", len(X_train))
print("Jumlah data testing:", len(X_test))
print("\nDistribusi training:", Counter(y_train))
print("Distribusi testing:", Counter(y_test))
```

```python
from sklearn.model_selection import train_test_split

# Split 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42
)

print("Jumlah data training:", len(X_train))
print("Jumlah data testing:", len(X_test))
print("\nDistribusi training:", Counter(y_train))
print("Distribusi testing:", Counter(y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Training model Naive Bayes
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi
print("=== HASIL EVALUASI NAIVE BAYES CLASSIFIER ===")
print(f"Akurasi  : {accuracy_score(y_test, y_pred)*100:.2f}%")
print(f"Precision: {precision_score(y_test, y_pred)*100:.2f}%")
print(f"Recall   : {recall_score(y_test, y_pred)*100:.2f}%")
print(f"F1-Score : {f1_score(y_test, y_pred)*100:.2f}%")

print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred, target_names=['Sedang', 'Tinggi']))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Sedang', 'Tinggi'],
            yticklabels=['Sedang', 'Tinggi'])
plt.title('Confusion Matrix - Naive Bayes Classifier')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
